```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('y')
plt.xlabel('x')
```

```python
plt.plot([1,3,4,9],[-1,8,1,2],'r')
```

```python
a = np.arange(2,5,0.2)
plt.plot(a, a**4, 'r--', a, a*100,'b--',)
```

```python
#c is with normal distribution
data = {
    'a' : np.arange(50),
    'b' : np.random.randint(0,50,50),
    'c' : np.random.randn(50)
}
plt.plot(data['a'], data['b'], 'r', data['a'], data['c'], 'b--')
```
